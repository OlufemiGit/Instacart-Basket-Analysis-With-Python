## Creating and working with Sampling Data

```python
# Creating Sampling Subset

df = ords_prods_merge[:2000000]
```

```python
np.random.seed(4)
```

```python
# Create a list holding True/False values to the test np.random.rant() <= 0.7

dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = ords_prods_merge[dev]
```

```python
# Store 30% of the sample in the dataframe small

small = ords_prods_merge[~dev]
```

```python
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
df_2 = small[['orders_day_of_week','prices']]
```

```python
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting the small sample line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Small_sample_line_chart_orders_dow.png'))
```
