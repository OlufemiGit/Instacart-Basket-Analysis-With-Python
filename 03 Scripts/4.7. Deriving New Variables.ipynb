## 03. Defining a Function

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range' ] = df.apply(price_label, axis=1)
```

```python
# Check the values of the new function
df['price_range']. value_counts(dropna = False)
```

```python
# Check and return max value of prices
df['prices'].max()
```

## 04. IF-Statement with the loc() Function

```python
# Creating a 'busiest day' column that shows busiest day, least busy, and regularly busy days.
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

Create a new column for this using a suitable method.

```python
# check which days are busiest and slowest

ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating the two busiest days and two slowest days column

result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value ==1:
    result.append("Busiest days")
  elif value == 3 or value== 4:
    result.append("Slowest days")
  else:
    result.append("Regularly busy")
```

```python
result
```

## When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
#Checking the columns' names
print(ords_prods_merge.columns)
```

```python
# Getting the value counts for each hour of day:

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create an empty list and divide the hours of the day into 3 sections starting with the most orders

result = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 12:
    result.append("Most orders")
  elif value == 17 or value == 14 or value == 15 or value == 13 or value == 16:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
result
```

```python
# Naming the column

ords_prods_merge['busiest_period_of_day'] = result
```

```python
# Printing the frequency for this new column

ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
