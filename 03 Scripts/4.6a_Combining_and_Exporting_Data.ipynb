## **05. Merging the dataframes**

###   **Merging Assessment:**
####  To combine the ***orders_products_combined*** dataframe with the ***df_prods*** dataframe, the best approach is to use a merge operation on the ***product_id column***, as it is the common key between both dataframes. The suitable method would be an inner join to retain only the matching records and save scarce memory resources.

```python
# Testing the merge without overwriting with merge flag: **'match'**

pd.merge(orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match')
```

### Assessment:

#### To improve the clarity and presentation of the merged dataset, the 'merge' column, which is a leftover from the previous merge, should be dropped after merging 'orders_products_combined' and 'df_prods'


## Merging dataframes, Dropping flag column '_merge' and confirming flag '_merge'

```python
# Actual merging of 'orders_products_combined' and '_df_prods' dataframes

ords_prods_merge = orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'merge_' )
```
