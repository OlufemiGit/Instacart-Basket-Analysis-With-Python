## Task 5:  Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day using a sample dataframe

## Task 6: Determine whether there’s a connection between age and family situation

```python
# Reduce previously made sample to the columns needed to make this determination
df_3 = small [['Age','Number of Dependents']]
```

```python
# Check to ensure the subset of the sample was made properly
df_3.head(5)
```

```python
# Create a line chart
line_2 = sns.lineplot(data = df_3, x = 'Age',y = 'Number of Dependents')
```

## Key Insight:
 - ***The chart shows that, on average, the number of dependents doesn't change much as people get older. There's some variation, but it's not a strong or predictable pattern.  This suggests that other things besides age are more important in determining how many dependents someone has.***

## Task 7: Create a scatter plot using the sns.scatterplot() function to explore whether there’s a connection between age and spending power (income)

```python
# Reduce previously made sample to only the columns needed
df_4 = small [['Age','Income']]
```

```python
# Create a scatterplot
scatterplot = sns.scatterplot(x = 'Age', y = 'Income', data = df_4)
```

## Insight:

  - ***The scatterplot shows that there are two main groups of people: those with lower incomes and those with higher incomes.  Getting older doesn't necessarily mean you'll earn more money. Something else, like your job or education, probably has a bigger impact on your income.***
